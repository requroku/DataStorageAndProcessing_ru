Помогите Василию выбрать три лучших варианта по совокупности параметров. Для этого надо применить **экспоненциальную нормировку**, сложить полученные нормированные значения параметров и выбрать лучшие варианты.

<sub>***В ответе укажите идентификаторы трех лучших вариантов через запятую и без пробелов, например: 3,10,7***</sub>

```python
import numpy as np
import pandas as pd
df = pd.read_csv('./data/data.csv')
#df.info()
df
```

```python
# Normalize values in the following columns
df['NORM_DISTANCE'] = 1 - np.exp(1 - df['DISTANCE'] / df['DISTANCE'].min())
df['NORM_STOP_COUNT'] = 1 - np.exp(1 - df['STOP_COUNT'] / df['STOP_COUNT'].min())
df['NORM_COST'] = 1 - np.exp(1 - df['COST'] / df['COST'].min())

# Create RESULT column as sum of normalized DISTANCE, STOP_COUNT, COST columns
df['RESULT'] = df['NORM_DISTANCE'] + df['NORM_STOP_COUNT'] + df['NORM_COST']
df
```

```python
# We are interested in the smallest DISTANCE from home to subway, the smallest number of stations
# from home to universuty and the lowest COST for the flat. That is why we look for the smallest values.
best_variants = df.nsmallest(3, 'RESULT')
best_variants
```

```python
# Print only IDs of the best variants
print(best_variants['ID'].values.tolist())
```
