## Теория
Особенность метода экспоненциального сглаживания заключается в том, что в процедуре нахождения сглаженного уровня используются значения только предшествующих уровней ряда, взятые с определенным весом, причем вес уменьшается по мере удаления его от момента времени, для которого определяется сглаженное значение уровня ряда. Если для исходного временного ряда $y_1, y_2, y_3,…, y_n$ соответствующие сглаженные значения уровней обозначить через $S_t, t = 1,2,...,n$, то экспоненциальное сглаживание осуществляется по формуле:  

$$S_t = (1-\alpha)y_t + \alpha S_{t-1}$$

В некоторых источниках приводится другая формула (**будет использоваться в нашем случае**):

$$S_t = \alpha y_t + (1-\alpha)S_{t-1}$$

где $\alpha$ - параметр сглаживания $(0 < \alpha < 1)$; величина $(1-\alpha)$ называется коэффициентом дисконтировании.

В практических задачах обработки экономических временных рядов рекомендуется (необоснованно) выбирать величину параметра сглаживания в интервале от $0.1$ до $0.3$. Других точных рекомендаций для выбора оптимальной величины параметра α пока нет. В отдельных случаях предлагается определять величину $\alpha$ исходя из длины сглаживаемого ряда: $\alpha = \frac{2}{(n+1)}$.
Что касается начального параметра $S_0$, то в задачах его берут или равным значению первого уровня ряда $у_1$, или равным средней арифметической нескольких первых членов ряда. Если при подходе к правому концу временного ряда сглаженные этим методом значения при выбранном параметре $\alpha$ начинают значительно отличаться от соответствующих значений исходного ряда, необходимо перейти на другой параметр сглаживания. Достоинством этого метода является то, что при сглаживании не теряются ни начальные, ни конечные уровни сглаживаемого временного ряда.

## Задача
Задан временной ряд $Y$ из $20$ элементов:  

$$Y = [40, 45, 46, 36, 35, 43, 42, 55, 53, 52, 54, 51, 62, 61, 59, 63, 47, 49, 68, 60]$$  

Ряд содержит шумы. Проведите экспоненциальное сглаживание временного ряда. При сглаживании используйте коэффициент сглаживания $\alpha = 0.25$.

Укажите сглаженное значение для последней (20-й) точки ряда:  
<sub>***При вводе ответа значение округлите до целого.***</sub>

```python
# Time series
Y = [40, 45, 46, 36, 35, 43, 42, 55, 53, 52, 54, 51, 62, 61, 59, 63, 47, 49, 68, 60]

# Coefficient
alpha = 0.25

# For which point of series calculate the value
point = 20

print('point \t Y \t St \t St (rounded)')

# Take mean value of first 3 elements as S_0
S_t = (Y[0] + Y[1] + Y[2]) / 3

# Take y_0 as S_0
#S_t = Y[0]

for t in range(len(Y)):
    # First formula (not correct answer for our case)
    #S_t = (1 - alpha)*Y[t] + alpha*S_t
    # Second formula - correct answer
    S_t = alpha*Y[t] + (1-alpha)*S_t
    print(f"{t+1} \t {Y[t]} \t {'{:.4f}'.format(S_t)} \t {round(S_t)}")
```
